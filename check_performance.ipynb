```python
import numpy as np
import json
import plotly.graph_objects as go
```

```python
with open('models/bad_run_4/transformer-plotting_type=mu-family=bert-n_positions=64-pretraining_sigma_tasks=1-pretraining_mu_tasks=1-default_sigma=0.2-family=bert-hid_dim=256-layer=4-head=4-tokenizer=learnable/plot_data.json') as json_file:
    data = json.load(json_file)

prediction = np.array(data["100000"]["prediction"])
x = np.array(data["100000"]["x"])
y = np.array(data["100000"]["y"])

print(x)
```

```python
print(prediction)
print(x)
print(y)
```

```python
fig = go.Figure()
x_axis = np.arange(prediction.shape[0])

fig.add_trace(go.Scatter(x=x_axis, y=prediction, mode='lines', name='prediction'))
fig.add_trace(go.Scatter(x=x_axis, y=x, mode='lines', name='x'))
fig.add_trace(go.Scatter(x=x_axis, y=y, mode='lines', name='mu'))
fig.update_layout(title='Comparing the model prediction, the true mu, and x (mu with noise, sigma=0.5)', xaxis_title='x', yaxis_title='y')

# make the plot into a scatter plot, not a line graph
# fig.update_traces(mode='markers', marker_line_width=1, marker_size=6)

fig.show()
```
